#### Código 1 - Para revisão de python

##### Estrutura sequencial

1. Fazer um aplicativo de terminal que leia um vaor inteiro correspondente à idade de uma pessoa em dias e informe-a e anos, meses e dias. Obs: Apenas para facilitar o cálculo, considere todo ano com 365 dias e todo mês com 30 dias. Nos casos de teste nunca averá uma situação que permite 12 meses e alguns dias como 360, 363, 364

```python
# Código 1 - Idade em Anos, Meses e Dias

#---- Entrada de dados -------#
dias = int(input("Digite a quantidade de dias: "))

#----Processamento de dados-----#
anos = dias // 365 # Calculo para saber a quantidade de anos
meses = (dias % 365) // 30 # Meses
dias = (dias % 365) % 30 # Dias restantes

#----Saída de Dados----#
print(f"Você possui {anos} anos, {meses} meses e {dias} dias de vida.")
```

#### match case

```python
codigo = int(input("Digite um código de 1 a 5: "))

match (codigo):
    case 1:
        print("Clodoaldo")
    case 2:
        print("SENAI")
    case 3:
        print("Americana")
    case 4:
        print("Campeão")
    case 5:
        print(codigo)
    case _:
        print("Nenhum valor válido digitado")
print("Fim pogam!")
```

#### Exemplo 1

```python
n = int(input("Digite um número de 1 a 10: "))

for i in range(0,10):
    print(f"{n} x {i + 1} = {n * (i+1)}")
```

#### Exemplo 2

```python
# Entrada de Dados
n = int(input("Quantos números você vai digitar? R: "))
# Processamento de dados
for i in range(n):
    numero = int(input("Digite um número: "))
    # Verificar se é IMPAR ou PAR
    if numero < 0:
        print("NEGATIVO")
    elif numero > 0:
        print("POSITIVO")
    else:
        print("NULO")
        continue
    if (numero % 2 == 0):
        print("PAR")
    elif (numero % 2 != 0):
        print("IMPAR")
```

### Estrutura de controle de repetição - Indeterminada

```python
while (True):
    senha = input("Digite a senha: ")
    if (senha == "2014"):
        print("Acesso permitido!")
        break
    else:
        print("Acesso negado...")
```

#### Exemplo

```python
while (True):
    print("Digite dois números: ")
    x = int(input())
    y = int(input())
    if x > y:
        print("Descrescente!")
    elif y > x:
        print("Crescente!")
    else:
        print("Iguais!")
        break
```

### Dicionário
 -> É uma coleção ordenada e mutável. Não permite membors duplicados. Cada item é um par chave-valor

```python
dicionario = {"SENAI": "5.07", "AMERICANA": 10, "CLODOALDO": True}
```

```python
print(dicionario["SENAI"])
```

```python
for chave in dicionario.keys():
    print(chave, "->", dicionario[chave])
```

```python
dicionario.update({"novo": "valor"})
print(dicionario)
```

#### Exemplo

```python
produtos = {
    1 : 5.00,
    2 : 3.50,
    3 : 4.80,
    4 : 8.90,
    5 : 7.32
}

codigo = int(input("Digite o código do produto: "))
quantidade = int(input("Digite a quantidade de produto a ser comprada: "))
try:
    print(f"O valor total da compra é de {produtos[codigo] * quantidade}")
except KeyError:
    print("Valor inválido. Use 1 a 5.")
```

> Se os parâmetros da função possuirem um valor padrão definido, é possível chamá-la somente com um ou menos argumentos.

```python
def deltaBhaskara(a,b,c):
    delta = (b * b) - (4*a*c)
    return delta
```
